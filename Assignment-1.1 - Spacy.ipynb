```
textData = sc.textFile('/FileStore/tables/The_Project_Gutenberg_eBook_of_Illustrat.txt')
```

```
textData.take(10)
```

```
punc = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
def remove_punc(string):
    for ele in string:
        if ele in punc:
            string = string.replace(ele, "")
    
    return string.lower()
```

```
textData = textData.map(lambda x: remove_punc(x))
```

```
text = ""
for s in textData.collect():
    text += s
```

```
text
```

```
spacy_model = en_core_web_sm.load()
```

```
entity_doc = spacy_model(text)
```

```
entity_doc.ents
```

```
mapData = [(entity.text, entity .label_) for entity in entity_doc.ents]
```

```
named = sc.parallelize(mapData)
```

```
named.take(10)
```

```
named = named.map(lambda x: (x[0].strip(), x[1]))
```

```
named.take(10)
```

```
named = named.map(lambda x : x[0]).map(lambda x: (x,1))
```

```
named.take(10)
```

```
named = named.reduceByKey(lambda x,y: x+y)
```

```
mostFrequentNamed = named.sortBy(lambda x: -x[1])
```

```
mostFrequentNamed.collect()
```
